**Import** **libraries**

```python
# Importing Libraries....
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load Dataset # read the dataset using pandas"read_csv"method
path ="/content/globalterrorismdb_0718dist.csv"
df= pd.read_csv(path,encoding = "ISO-8859-1")
```

**Data** **Cleaning**

```python
# Data cleaning using drop function
df.drop('eventid',axis=1,inplace = True)
```

```python
#checking first row of dataframe after data cleaning
df.head(1)
```

```python
#checking the shape of dataframe
df.shape
```

```python
#checking null values of dataset
pd.DataFrame(df.isna().sum()).T
```

```python
pd.DataFrame(df.isnull().sum()).T
```

```python
#dropping values if a column have 50% of null value
thresold = len(df)*0.5
df = df.dropna(thresh= thresold,axis=1)
```

```python
pd.DataFrame(df.isnull().sum()).T
```

```python
df.shape  # checking shape
```

```python
#covert columns into list 
list_col = df.columns.to_list()
```

```python
#getting description of columns sepratly
for i in list_col:
  print(df[i].describe())
  print("-------------")
```

```python
#selecting column and storing float type elements in list
list_num_cols = df.select_dtypes(include = ['float']).columns.to_list()
```

```python
for i in list_num_cols:   # filling na value with mean of column
  df[i].fillna(df[i].mean(),inplace = True)
```

```python
#storing all the columns having object data types
list_str_cols = df.select_dtypes(include = ['object']).columns.to_list()
```

```python
# replace all the na values for mode
for i in list_str_cols:
  df[i].fillna(df[i].mode(),inplace = True)
```

```python
df.head(1) # again checking na value
```

## ***4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables***

```python
corr =df.corr() # heatmap for given data
sns.heatmap(corr) 
```

```python
# Chart - 1 visualization code plotting bar graph
sns.set_theme(style='whitegrid')
plt.figure(figsize =(24,9))
plt.ylim(0,2000,500)
plt.xticks(rotation=90)
sns.countplot(x=df['iyear'],palette="Set3")
```

**Inferences:**

AS we can see mostly numbers of attacks in 2014 than in 2015 and 2016.

```python
# diffrenciate the bar graph using function 'hue'.and finding duration of attacks
sns.set_theme(style ="whitegrid")
plt.figure(figsize=(22,10))
plt.ylim(0,20000,500)
plt.xticks(rotation =90)
sns.countplot(x=df['iyear'], hue = df['extended'])
```

This chart represents the how long terrorist attack is going on.


Mostly extended attacks were in 2014 and 2015.


```python
#number of attacks according to months
sns.set_theme(style="whitegrid")
plt.figure(figsize=(22,10))
plt.ylim(0,20000,500)
plt.xticks(rotation=90)
plt.ylabel('Number of attacks')
plt.xlabel('Months')
sns.countplot(x=df['imonth'],palette ="Set3")
```

Highest number of attacks were in  fifth(may) month.


this chart represnts the number of attacks according to months.

```python
#diffrenciate the number of attacks in month chart
sns.set_theme(style="whitegrid")
plt.figure(figsize=(22,10))
plt.ylim(0,20000,400)
plt.xticks(rotation=90)
plt.ylabel('Number of attacks')
plt.xlabel('months')
sns.countplot(x=df['imonth'],hue = df['extended'])
```
